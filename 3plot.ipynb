```python
tips = sns.load_dataset('tips')
print(tips.head())
tips.shape
```

```python
#hist
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
axes1.hist(tips['total_bill'],bins=10)
axes1.set_title('histogram of total bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
```

```python
#scatterplot
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip'])
axes1.set_title('Scatterplot of total bill vs tips')
axes1.set_xlabel('Total bill')
axes1.set_ylabel('Tip')
```

```python
#boxplot
boxplot=plt.figure()
axes1=boxplot.add_subplot(1,1,1)
axes1.boxplot(
    #1st arg of boxplot is the data, since we plot multipy data, we need to put each piece of data into a list
    [tips[tips['sex']=='Female']['tip'],
    tips[tips['sex']=='Male']['tip']],
    labels=['Female','Male']
    )
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

```python
#multivariate data
#creat a color variable based on sex
def recode_sec(sex):
    if sex =='Female':
        return 0
    return 1

tips['sex_color']=tips['sex'].apply(recode_sec)
print(tips.head())
```

```python
scatter_plot = plt.figure()
axes1=scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    #set the size of dots based on party size and we multiply the values by 10 to make the points bigger to emphasize the diff
    s=tips['size']*10,
    #set the alpha value so points are more transparent
    alpha=0.5)
axes1.set_title('total bill vs tip colored by sex and sized by size')
axes1.set_xlabel('total bill')
axes1.set_ylabel('tip')
```

```python
#3.4Seaborn
import seaborn as sns

tips = sns.load_dataset('tips')
#univariate
#histogram
hist,ax=plt.subplots()

#use the distplot function from seaborn to create our plot
ax=sns.distplot(tips['total_bill'])
ax.set_title('Total bill histogram with density plot')
```

```python
#w/o density
hist,ax=plt.subplots()
ax=sns.distplot(tips['total_bill'],kde=False)
ax.set_title('Total bill histogram without density plot')
ax.set_xlabel('Total bill')
ax.set_ylabel('Frequency')
```

```python
#w/o hist
den,ax=plt.subplots()
ax=sns.distplot(tips['total_bill'],hist=False)
ax.set_title('Total bill density plot')
ax.set_xlabel('Total bill')
ax.set_ylabel('unit Probability')
```

```python
#rug plot
hist_den_rug,ax=plt.subplots()
ax=sns.distplot(tips['total_bill'],rug=True)
ax.set_title('Total bill histtogrram with density and rug plot')
ax.set_xlabel('Total bill')
```

```python
#count plot/bar plot
count,ax=plt.subplots()
ax=sns.countplot('day',data=tips)
ax.set_title('count of the day')
ax.set_xlabel('day of the week')
ax.set_ylabel('frequency')
```

```python
#bivariate data
#scatter plot
scatter,ax=plt.subplots()
ax=sns.regplot(x='total_bill', y='tip', data=tips)
ax.set_title('scatterplot of the total bill and tip')
ax.set_xlabel('total bill')
ax.set_ylabel('tip')
```

```python
#similar
fig = sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
joint=sns.jointplot(x='total_bill',y='tip',data=tips)
joint.set_axis_labels(xlabel='total bill',ylabel='tip')
#add title, set font size and move the text above the total bill axe
joint.fig.suptitle('joint plot of total bill and tip',fontsize=10,y=1.03)
```

```python
#hexbin plot
hexbin = sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
hexbin.set_axis_labels(xlabel='total bill',ylabel='tip')
hexbin.fig.suptitle('hexbin joint plot of thetotal; bill and tip',
                   fontsize=10,y=1.05)
```

```python
#2d density plot 
kde,ax=plt.subplots()
ax=sns.kdeplot(data=tips['total_bill'],
              data2=tips['tip'],
              shade=True)
ax.set_title('kenel density plot of total bill and tip')
```

```python
kde_joint=sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
#bar plot
bar,ax =plt.subplots()
ax= sns.barplot(x='time',y='total_bill',data=tips)
ax.set_title('bar plot of avgtotal bill for time of the day')
```

```python
#boxplot
box,ax=plt.subplots()
ax= sns.boxplot(x='time',y='total_bill',data=tips)
```
